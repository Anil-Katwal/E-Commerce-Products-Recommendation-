```python
import pandas as pd

def clean_text(row):
    title = row["title"] if pd.notna(row["title"]) else ''
    brand = row["brand"] if pd.notna(row["brand"]) else ''
    category = row["category"] if pd.notna(row["category"]) else ''
    description = row["description"] if pd.notna(row["description"]) else ''

    specs = ''
    if pd.notna(row["specifications"]):
        try:
            specs_dict = eval(row["specifications"])
            if isinstance(specs_dict, dict):
                specs = ' '.join([f"{k}:{v}" for k, v in specs_dict.items()])
        except:
            specs = ''

    return f"{title} {brand} {category} {description} {specs}"
```

```python
df['clean_text'] = df.apply(clean_text, axis=1)
df = df.dropna(subset=['clean_text'])
```

```python
from sentence_transformers import SentenceTransformer
model=SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
df['embeddings'] = df['clean_text'].apply(lambda x: model.encode(x))
```

## Recommednation Engine

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity

def recommend_product(query, top_k=5):
    # Encode the query
    query_embedding = model.encode(query)

    # Compute similarity
    df['similarity'] = df['embeddings'].apply(
        lambda x: cosine_similarity([query_embedding], [x])[0][0]
    )

    # Sort and get top_k recommendations
    recommendations = df.sort_values(by='similarity', ascending=False).head(top_k)

    return recommendations[['title', 'brand', 'category', 'similarity']]

# Example usage:
query = '8GB RAM smartphone'
recommendations = recommend_product(query)
recommendations
```
